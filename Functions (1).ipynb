1.  What is the difference between a function and a method in Python?

-> Function is a standalone and reusable block of code that performs a specific task whereas method is a function that is associated with an object.

-> A function is defined using the def keyword outside a class, whereas a method is also defined using the def keyword but inside a class.

-> Function can be called independently, without being associated with any object, whereas a method is called on an object (instance of a class) and operates on that specific instance.

-> Functions may or may not take arguments and return values whereas the first parameter of a method is typically self, which represents the instance of the class
   

```python
# Example

# Defining a Function
def greet(name):
  return f"Hello, {name}!"

# creating Method inside class
class Person:
  def __init__(self, name):
    self.name = name

  def introduce (self):
    return f"My name is {self.name}."

#calling the function
print(greet("Alice"))

# Creating an object of the class
person1 = Person("Alice")

# Calling the method using the object
print(person1.introduce())
```

5. What are iterators in Python and how do they differ from iterables?

->  An iterator is an object that represents a stream of data and returns elements one by one. An iterable is any Python object, sequential structure, or data structure that can return its members one at a time, allowing it to be iterated over using a for loop or a while loop.







```python
# Iterable (list)
my_list = [1, 2, 3]

# Creating an iterator from the iterable
iter_obj = iter(my_list)

# Using next() to get elements one by one
print(next(iter_obj))  # Output: 1
print(next(iter_obj))  # Output: 2
print(next(iter_obj))  # Output: 3

# Calling next(iter_obj) again will raise StopIteration as all elements are exhausted
```

6.  Explain the concept of generators in Python and how they are defined.

-> A generator is a special type of iterator that simplifies the process of creating iterators. Instead of using __iter__() and __next__(), generators are defined using functions with the yield keyword. They produce values one at a time, pausing execution and resuming from where they left off when next() is called.

```python
#example

def count_up_to(n):
    count = 1
    while count <= n:
        yield count  # Pauses execution and returns value
        count += 1

# Creating a generator object
gen = count_up_to(3)

# Fetching values using next()
print(next(gen))  # Output: 1
print(next(gen))  # Output: 2
print(next(gen))  # Output: 3

# Calling next(gen) again will raise StopIteration as all elements are exhausted
```

7. What are the advantages of using generators over regular functions?

-> A generator function generates results one by one instead of returning all results at once. This makes it memory-efficient and reduces execution time, as values are produced on demand rather than stored in memory.

```python
#example regular function

def squares_list(n):
    result = []
    for i in range(n):
        result.append(i ** 2)
    return result

print(squares_list(5))  # Output: [0, 1, 4, 9, 16]
```

```python
#generator function

def squares_generator(n):
    for i in range(n):
        yield i ** 2  # Generates values one by one

gen = squares_generator(5)

print(next(gen))  # Output: 0
print(next(gen))  # Output: 1
print(next(gen))  # Output: 4
```

10. What is the difference between `map()`, `reduce()`, and `filter()` functions in Python?

-> The map(), filter(), and reduce() functions in Python are used for functional programming, each serving a distinct purpose. The map() function applies a given function to each element in an iterable and returns an iterator with transformed values. It is useful when modifying all elements in a sequence, such as squaring a list of numbers. The filter() function, on the other hand, is used to filter elements based on a specific condition, returning only those that satisfy it. This is ideal for selecting values from a dataset, like extracting even numbers from a list. Lastly, the reduce() function (from the functools module) cumulatively applies a function to elements in an iterable, reducing them to a single value. It is commonly used for operations like summing all numbers in a list. While both map() and filter() return iterators, reduce() produces a single value. These functions help write concise, efficient, and readable code, avoiding the need for manual loops.

-> Syntax

a. map(function, iterable)


b. filter(function, iterable)

c. from functools import reduce

reduce(function, iterable)



```python
#map
nums = [1, 2, 3, 4]
squared = list(map(lambda x: x ** 2, nums))
print(squared)  # Output: [1, 4, 9, 16]


#filter
nums = [1, 2, 3, 4, 5, 6]
evens = list(filter(lambda x: x % 2 == 0, nums))
print(evens)  # Output: [2, 4, 6]


#reduce
from functools import reduce
nums = [1, 2, 3, 4]
sum_all = reduce(lambda x, y: x + y, nums)
print(sum_all)  # Output: 10
```

```python
''' 4. Write a Python function that checks if a given number is prime or not from 1 to 200. '''

def is_prime(n):
    if n < 2:
        return False  # 0 and 1 are not prime numbers
    for i in range(2, int(n ** 0.5) + 1):  # Check divisibility up to sqrt(n)
        if n % i == 0:
            return False  # If divisible, not prime
    return True  # If not divisible by any number, it's prime

# Print prime numbers from 1 to 200
prime_numbers = [num for num in range(1, 201) if is_prime(num)]
print(prime_numbers)

```

```python
'''5. Create an iterator class in Python that generates the Fibonacci sequence up to a specified number of terms. '''

class FibonacciIterator:
    def __init__(self, n_terms):
        self.n_terms = n_terms  # Number of terms to generate
        self.a, self.b = 0, 1   # Initialize first two Fibonacci numbers
        self.count = 0          # Counter to track iterations

    def __iter__(self):
        return self  # An iterator must return itself

    def __next__(self):
        if self.count >= self.n_terms:
            raise StopIteration  # Stop iteration when limit is reached

        if self.count == 0:  # First Fibonacci number
            self.count += 1
            return self.a
        elif self.count == 1:  # Second Fibonacci number
            self.count += 1
            return self.b
        else:
            fib = self.a + self.b  # Calculate next Fibonacci number
            self.a, self.b = self.b, fib  # Update values
            self.count += 1
            return fib

# Example usage
fib_iter = FibonacciIterator(10)  # Generate first 10 Fibonacci numbers

for num in fib_iter:
    print(num, end=" ")
```

```python
''' 7.  Implement a generator function that reads a file line by line and yields each line as a string. '''

def read_file_line_by_line(file_path):
    """Generator function to read a file line by line."""
    with open(file_path, 'r', encoding='utf-8') as file:
        for line in file:
            yield line.strip()  # Yield each line after stripping leading/trailing whitespace

# Example usage
file_path = "sample.txt"  # Replace with your actual file path

for line in read_file_line_by_line(file_path):
    print(line)  # Prints each line one by one
```

```python
'''9.  Write a Python program that uses `map()` to convert a list of temperatures from Celsius to Fahrenheit. '''

# Function to convert Celsius to Fahrenheit
def celsius_to_fahrenheit(celsius):
    return (celsius * 9/5) + 32  # Formula: (C  9/5) + 32 = F

# List of temperatures in Celsius
celsius_temperatures = [0, 10, 20, 30, 40, 100]

# Using map() to apply the conversion function to each element in the list
fahrenheit_temperatures = list(map(celsius_to_fahrenheit, celsius_temperatures))

# Printing the result
print(fahrenheit_temperatures)
```

```python
'''10.  Create a Python program that uses `filter()` to remove all the vowels from a given string. '''

def is_not_vowel(char):
    """Returns True if the character is NOT a vowel"""
    return char.lower() not in "aeiou"

def remove_vowels(string):
    """Removes vowels from a given string using filter()"""
    return "".join(filter(is_not_vowel, string))

# Example usage
input_string = "Hello, World!"
result = remove_vowels(input_string)

print(result)  # Output: "Hll, Wrld!"
```
